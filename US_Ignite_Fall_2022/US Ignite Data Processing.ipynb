```python
def ChangeTimeToMST(value):
    x = value
    x = x.replace('UTC', '').strip()

    utc = datetime.strptime(x, '%Y-%m-%d %H:%M:%S')
    utc = utc.replace(tzinfo=tz.gettz('UTC'))

    mtc = utc.astimezone(pytz.timezone('MST'))
    mtc = str(mtc)
    mtc = mtc[:13]
    mtc = datetime.strptime(mtc, '%Y-%m-%d %H')
    mtc = str(mtc)
    mtc = mtc.replace(" ", "")
    mtc = mtc.replace(":", "")
    mtc = mtc.replace("-", "")
    result = mtc[:10]
    
    return result
```
