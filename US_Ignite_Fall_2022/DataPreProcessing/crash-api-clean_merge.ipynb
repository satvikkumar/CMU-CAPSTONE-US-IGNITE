```python
from google.colab import files
import io
import pandas as pd
uploaded1 = files.upload()
clean_weather = pd.read_csv(io.BytesIO(uploaded1['WeatherData20years - WeatherData20years.csv']))
```

```python
clean_weather.head()
```

```python
clean_weather['dt_iso'] = clean_weather['dt_iso'].str.split('+').str.get(0)
clean_weather['dt_iso'] = pd.to_datetime(clean_weather['dt_iso'])
clean_weather['ts'] = clean_weather['dt_iso'].dt.strftime('%Y-%m-%d %H:%M')
clean_weather
```

```python
clean_weather['ts'] = pd.to_datetime(clean_weather['ts'])
clean_weather['ts'].dtype
```

```python
weather_2019 = clean_weather[(clean_weather['dt_iso']>='2019-01-01') & (clean_weather['dt_iso']<='2019-12-31')]
weather_2021 = clean_weather[(clean_weather['dt_iso']>='2021-01-01') & (clean_weather['dt_iso']<='2021-12-31')]
weather_1921 = weather_2019.append(weather_2021, ignore_index=True)
weather_1921
```

```python
waze_2019 = clean_waze[(clean_waze['ts']>='2019-01-01') & (clean_waze['ts']<='2019-12-31')]
waze_2021 = clean_waze[(clean_waze['ts']>='2021-01-01') & (clean_waze['ts']<='2021-12-31')]
waze_1921 = waze_2019.append(waze_2021, ignore_index=True)
```

```python
clean_waze
```

```python
waze_1921 = waze_1921.iloc[:,1:]
waze_1921 = waze_1921.drop_duplicates()
```

```python
waze_1921
```

```python
waze_1921 = waze_1921.sort_values(by=['ts'])
```

```python
waze_1921_median = waze_1921.groupby(['street','ts']).agg (median_speed= ('speed', 'median'), average_jam_level=('level','median'))
waze_1921_median = waze_1921_median.reset_index()
waze_1921_median
```

```python
weather_1921['ts'] = pd.to_datetime(weather_1921['ts'])
waze_1921_median['ts'] = pd.to_datetime(waze_1921_median['ts'])
```

```python
weather_waze_1921 = pd.merge(weather_1921,waze_1921_median,on='ts') 
weather_waze_1921
```

```python
clean_crash['crash_label'] = 1
clean_crash['ts'] = pd.to_datetime(clean_crash['ts'])
```

```python
merge_tmp = pd.merge(weather_waze_1921, clean_crash, how='left', on=['street','ts'])
merge_tmp['crash_label'] = merge_tmp['crash_label'].fillna(0)
merge_tmp.groupby(['crash_label']).count()
```
